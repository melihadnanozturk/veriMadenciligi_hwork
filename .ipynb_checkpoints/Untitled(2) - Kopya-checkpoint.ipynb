```python
kmeans = KMeans()   # boÅŸ bir KMeans nesnesi oluÅŸturma
ssd = []            # ssd isimli bos bir liste oluÅŸturma
K = range(1, 30)    # 1'den 30'a kadar K deÄŸerlerini oluÅŸturma

# DÃ¶ngÃ¼ aracÄ±lÄ±ÄŸÄ±yla iÅŸlemleri gerÃ§ekleÅŸtirme
for k in K:
    kmeans = KMeans(n_clusters = k).fit(df_normal)
    ssd.append(kmeans.inertia_)

# ssd deÄŸerlerini gÃ¶rselleÅŸtirme
plt.plot(K, ssd)
plt.xlabel('K deÄŸerleri')
plt.ylabel('SSD')
plt.title('Elbow Methodu ile SSD deÄŸerlerini gÃ¶rselleÅŸtirme')
plt.show()
```

```python
x = df_normal.copy()
kmeans = kmeans = KMeans(n_clusters=5, random_state=0).fit(x)
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
```

```python
pca = PCA(n_components=2)
df_pca = pca.fit_transform(x)
```

```python
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.5, label='Veriler')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', label='KÃ¼me Merkezleri')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('K-Means ile KÃ¼meleme')
plt.legend()
plt.show()
```

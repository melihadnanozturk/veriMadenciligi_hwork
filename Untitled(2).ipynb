```python
kmeans = KMeans()   # boÅŸ bir KMeans nesnesi oluÅŸturma
ssd = []            # ssd isimli bos bir liste oluÅŸturma
K = range(1, 30)    # 1'den 30'a kadar K deÄŸerlerini oluÅŸturma

# DÃ¶ngÃ¼ aracÄ±lÄ±ÄŸÄ±yla iÅŸlemleri gerÃ§ekleÅŸtirme
for k in K:
    kmeans = KMeans(n_clusters = k).fit(df)
    ssd.append(kmeans.inertia_)

# ssd deÄŸerlerini gÃ¶rselleÅŸtirme
plt.plot(K, ssd)
plt.xlabel('K deÄŸerleri')
plt.ylabel('SSD')
plt.title('Elbow Methodu Kategorili Veriler')
plt.show()
```

```python
def fit_kproto(n_cluster, data):
    kproto = KPrototypes(n_clusters=n_cluster, init='Cao', verbose=0.1)
    return kproto.fit_predict(data, categorical=[1, 3, 8, 12]), kproto
```

```python
clusters, _ = fit_kproto(5, df)
principalComponents = PCA(n_components=2).fit_transform(df)
principalDf = pd.DataFrame(data=principalComponents, columns=['PC1', 'PC2'])
principalDf['Cluster'] = clusters

sns.scatterplot(x="PC1", y="PC2", hue="Cluster", data=principalDf, palette="Set2")
plt.title('PCA Visualization of Sleep Health Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
